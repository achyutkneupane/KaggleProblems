```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Kaggle/Datasets/titanic/train.csv')
test = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Kaggle/Datasets/titanic/test.csv')
```

```python
train.head()
```

```python
# Check for missing values
print(train.isnull().sum())

# Check the distribution of the target variable
print(train['Survived'].value_counts())
```

```python
# Check for missing values
print(test.isnull().sum())
```

```python
train['Age'] = train['Age'].fillna(train['Age'].mean())
test['Age'] = test['Age'].fillna(test['Age'].mean())
test['Fare'] = test['Fare'].fillna(test['Fare'].mean())
```

```python
print(train.isnull().sum())
print(test.isnull().sum())
```

```python
train = train.drop('Cabin', axis=1)
test = test.drop('Cabin', axis=1)
```

```python
train.head()
```

```python
test.head()
```

```python
train['Embarked'] = train['Embarked'].fillna('S')
```

```python
print(train.isnull().sum())
print(test.isnull().sum())
```

```python
# Create a title variable from the name variable
train['Title'] = train['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)

# Create a family size variable by combining the siblings/spouse and parent/child variables
train['FamilySize'] = train['SibSp'] + train['Parch'] + 1

# Create a variable indicating whether the passenger is alone or not
train['IsAlone'] = np.where(train['FamilySize'] == 1, 1, 0)
```

```python
train.head()
```

```python
# Create a title variable from the name variable
test['Title'] = test['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)

# Create a family size variable by combining the siblings/spouse and parent/child variables
test['FamilySize'] = test['SibSp'] + test['Parch'] + 1

# Create a variable indicating whether the passenger is alone or not
test['IsAlone'] = np.where(test['FamilySize'] == 1, 1, 0)
```

```python
test.head()
```

```python
features = ["Pclass", "Sex", "Age", "SibSp", "Parch","Fare", "Embarked", "FamilySize", "IsAlone"]
X = pd.get_dummies(train[features])
y = train["Survived"]
X_test = pd.get_dummies(test[features])
```

```python
X.head()
```

```python
X_test.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.model_selection import cross_val_score

# Split the data into features and target variable
X_train = X
y_train = y

# Define a list of models to try
models = [LogisticRegression(), DecisionTreeClassifier(), RandomForestClassifier(), GradientBoostingClassifier()]

# Try each model and print its cross-validation score
for model in models:
    score = cross_val_score(model, X_train, y_train, cv=5).mean()
    print(f"{type(model).__name__} score: {score:.4f}")
```

```python
from sklearn.model_selection import GridSearchCV

# Define the hyperparameter grid for the random forest classifier
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7],
    'min_samples_split': [2, 4, 6]
}

# Instantiate the random forest classifier
rfc = RandomForestClassifier()

# Use grid search to find the best hyperparameters
grid_search = GridSearchCV(rfc, param_grid=param_grid, cv=5)
grid_search.fit(X_train, y_train)

# Print the best hyperparameters
print(grid_search.best_params_)
```

```python
from sklearn.ensemble import VotingClassifier

# Define the list of models to use in the voting classifier
models = [('lr', LogisticRegression()), ('dt', DecisionTreeClassifier()), ('rf', RandomForestClassifier(n_estimators=200, max_depth=5, min_samples_split=2)), ('gb', GradientBoostingClassifier())]

# Instantiate the voting classifier
voting_clf = VotingClassifier(models)

# Evaluate the voting classifier's performance using cross-validation
score = cross_val_score(voting_clf, X_train, y_train, cv=5).mean()
print(f"VotingClassifier score: {score:.4f}")
```

```python
voting_clf.fit(X,y)
```

```python
predictions = voting_clf.predict(X_test)
```

```python
output = pd.DataFrame({'PassengerId': test.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
```
